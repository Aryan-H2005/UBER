```python
import pandas as pd

df = pd.read_csv("uber_bookings.csv")

# ---------------------------
# 1. Target / KPI columns
# ---------------------------
kpi_cols = [
    'Booking Value',
    'Ride Distance',
    'Customer Rating',
    'Driver Ratings'
]

for col in kpi_cols:
    df[col] = df[col].fillna(0)

# ---------------------------
# 2. Categorical columns
# ---------------------------
cat_cols = [
    'Payment Method',
    'Reason for cancelling by Customer',
    'Driver Cancellation Reason',
    'Incomplete Rides Reason'
]

for col in cat_cols:
    df[col] = df[col].fillna("Unknown")

# ---------------------------
# 3. Boolean / Flag columns
# ---------------------------
flag_cols = [
    'Cancelled Rides by Customer',
    'Cancelled Rides by Driver',
    'Incomplete Rides'
]

for col in flag_cols:
    df[col] = df[col].fillna(0)

# ---------------------------
# 4. Export cleaned data
# ---------------------------
df.to_csv("uber_cleaned_powerbi.csv", index=False)
```
